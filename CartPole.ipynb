```python
import gym
import numpy as np
import time
%matplotlib inline
import matplotlib.pyplot as plt
import random
```

```python
env = gym.make("CartPole-v1")
print(env.observation_space)
print(env.action_space)
```

```python

class No_Learning_Agent():
    def __init__(self,env):
        self.act_size = env.action_space.n
        print("action size",self.act_size)
    
    def get_action(self,state):
        action = random.choice(range(self.act_size))                                                                                                                                       
        pole_angle = state[2]
        if pole_angle < 0:
            action = 0
        else:
            action = 1
        return action
```

```python
agent = No_Learning_Agent(env)

no_of_episodes = 100
rewards_list = []
for x in range(no_of_episodes):
    reward1 = 0
    state = env.reset()
    print(state)
    for i in range(200):
        #action = env.action_space.sample()
        action = agent.get_action(state)
        state, reward, done, info = env.step(action)
        print("step = ",i,", state = ",state)
        reward1 = reward1 + reward
        #env.render()
        #time.sleep(0.05)
        if done == True:
            #print (i)
            #print(state)
            #print(reward)
            #print(reward1)
            rewards_list.append(reward1)
            break
print("Average reward over ",no_of_episodes," episodes = ",np.average(rewards_list))
print("Max reward in an episode over ",no_of_episodes," episodes = ",np.max(rewards_list))
print("Min reward in an episode over",no_of_episodes," episodes = ",np.min(rewards_list))
    
```
